```python
import matplotlib.pyplot as plt
import math

# --- CONFIGURACIÓN INICIAL ---

# Funciones de robustez
funciones = ["gradopromedio", "LCC", "entropia1", "entropia2", "entropia3"]

# Modelos de redes
modelos = ["ba", "ws"]

# Escenarios de fallos/ataques
escenarios = ["random", "degree"]

# Diccionario para nombres bonitos
nombres = {
    "gradopromedio": "<k>",
    "LCC": "LCC",
    "entropia1": "H_1(G)",
    "entropia2": "H_2(G)",
    "entropia3": "H_3(G)",
    "ba": "BA",
    "ws": "WS",
    "random": "fallos aleatorios",
    "degree": "ataques por grado mayor"
}

# --- GENERACIÓN DE COMBINACIONES VÁLIDAS ---

plots_a_hacer = []

for funcion1 in funciones:
    for funcion2 in funciones:
        if funcion1 == funcion2:
            continue

        if funcion1 == "gradopromedio":
            if funcion2 == "LCC":
                continue
            tipo = "grado_vs_entropias"

        elif funcion1 == "LCC":
            if funcion2 == "gradopromedio":
                continue
            tipo = "lcc_vs_entropias"

        elif funcion1 == "entropia1":
            if funcion2 in ["gradopromedio", "LCC"]:
                continue
            tipo = "entropias_vs_entropias"

        elif funcion1 == "entropia2":
            if funcion2 in ["gradopromedio", "LCC", "entropia1"]:
                continue
            tipo = "entropias_vs_entropias"

        else:
            continue

        for escenario in escenarios:
            plots_a_hacer.append((funcion1, funcion2, escenario, tipo))

# --- SEPARAR POR ESCENARIO ---

plots_random = [p for p in plots_a_hacer if p[2] == "random"]
plots_degree = [p for p in plots_a_hacer if p[2] == "degree"]

# --- FUNCIÓN PARA GRAFICAR MATRIZ DE SUBPLOTS ---

def graficar_matriz(plots, titulo_figura, nombre_archivo):
    n = len(plots)
    cols = 3
    rows = math.ceil(n / cols)

    fig, axs = plt.subplots(rows, cols, figsize=(6 * cols, 4 * rows))
    axs = axs.flatten()

    for idx, (funcion1, funcion2, escenario, tipo) in enumerate(plots):
        ax = axs[idx]
        for modelo in modelos:
            var1_nombre = f"R_{funcion1}_{modelo}_{escenario}"
            var2_nombre = f"R_{funcion2}_{modelo}_{escenario}"

            try:
                y1 = globals()[var1_nombre]
                y2 = globals()[var2_nombre]
            except KeyError:
                print(f"variable {var1_nombre} o {var2_nombre} no encontrada")
                continue

            ax.plot(y1, label=f'{nombres[funcion1]} - {nombres[modelo]}')
            ax.plot(y2, label=f'{nombres[funcion2]} - {nombres[modelo]}')

        ax.set_title(f"{nombres[funcion1]} vs {nombres[funcion2]}", fontsize=10)
        ax.set_xlabel("k")
        ax.set_ylabel("R_k")
        ax.grid(True)
        ax.legend(fontsize=8)

    # Ocultar subplots vacíos
    for i in range(n, len(axs)):
        fig.delaxes(axs[i])

    fig.suptitle(titulo_figura, fontsize=16)
    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.savefig(nombre_archivo, format="pdf")
    plt.show()

# --- LLAMADAS PARA GRAFICAR ---

graficar_matriz(plots_random, "Escenario de Fallos Aleatorios", "plots_fallos.pdf")
graficar_matriz(plots_degree, "Escenario de Ataques por Grado", "plots_ataques.pdf")
```
