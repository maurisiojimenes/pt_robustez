```python
import graficas as g
import numpy as np
import entropia as en
import time
import matplotlib.pyplot as plt
import importlib
```

```python
G_BA = g.crear_grafica('barabasi',1000,15,0)
G_WS = g.crear_grafica('watts',1000,30,0.5)
```

```python
P_BA = g.obtener_distribucion_1(G_BA)
P_WS = g.obtener_distribucion_1(G_WS)


n_BA=[]#Lista que va a almacenar los nodos que se fueron quitando
n_WS=[]

G_1 = G_BA.copy()
G_2 = G_WS.copy()

N = len(G_BA)

DKL_BA=[]
DKL_WS=[]
inicio = time.time()

for i in range(N):
    print(f"Vamos en la iteración: {i}")
    G_1,nodo_BA = g.ataques(G_1,'random')
    G_2,nodo_WS = g.ataques(G_2,'random')
    n_BA.append(nodo_BA)
    n_WS.append(nodo_WS)
    Q_BA = g.obtener_distribucion_1(G_1)
    Q_WS = g.obtener_distribucion_1(G_2)
    dkl_BA = en.DKL_1(P_BA,Q_BA,n_BA)
    dkl_WS = en.DKL_1(P_WS,Q_WS,n_WS)
    print(dkl_BA,dkl_WS)
    DKL_BA.append(dkl_BA)
    DKL_WS.append(dkl_WS)
    
final = time.time()

print(f"El tiempo de ejecución fue de: {final - inicio}")
```

```python
x = np.linspace(1,100,N)

plt.plot(x,DKL_BA,label='Modelo - BA',color='red')
plt.plot(x,DKL_WS,label='Modelo - WS',color='lightblue')
plt.title('Divergencia de Kullback Leibler usando H_1')
plt.xlabel('% de nodos atacados')
plt.ylabel('DKL')
plt.legend()
plt.grid(True)
```
