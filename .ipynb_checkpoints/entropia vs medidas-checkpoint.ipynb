```python
import graficas as g
import entropia as en
import matplotlib.pyplot as plt
import numpy as np
import time
import importlib
```

```python
G_barabasi = g.crear_grafica('barabasi',500,10,0)
G_watts = g.crear_grafica('watts',500,10,0.5)

N = len(G_barabasi)

inicio = time.time()

entropia_1_barabasi = en.entropia_normalizada_1(G_barabasi)
entropia_1_watts = en.entropia_normalizada_1(G_watts)
entropia_2_barabasi = en.entropia_normalizada_2(G_barabasi)
entropia_2_watts = en.entropia_normalizada_2(G_watts)
GP_barabasi = g.grado_promedio(G_barabasi)
GP_watts = g.grado_promedio(G_watts)
CC_barabasi = g.componentes_conexas(G_barabasi)
CC_watts = g.componentes_conexas(G_watts)

entropias_1_barabasi = [entropia_1_barabasi]
entropias_1_watts = [entropia_1_watts]
entropias_2_barabasi = [entropia_2_barabasi]
entropias_2_watts = [entropia_2_watts]
GPs_barabasi = [GP_barabasi]
GPs_watts = [GP_watts]
CCs_barabasi = [CC_barabasi]
CCs_watts = [CC_watts]

for i in range(N):
    
    G_barabasi = g.ataques(G_barabasi,'random')
    G_watts = g.ataques(G_watts,'random')
    
    entropia_1_barabasi = en.entropia_normalizada_1(G_barabasi)
    entropia_1_watts = en.entropia_normalizada_1(G_watts)
    entropia_2_barabasi = en.entropia_normalizada_2(G_barabasi)
    entropia_2_watts = en.entropia_normalizada_2(G_watts)
    GP_barabasi = g.grado_promedio(G_barabasi)
    GP_watts = g.grado_promedio(G_watts)
    CC_barabasi = g.componentes_conexas(G_barabasi)
    CC_watts = g.componentes_conexas(G_watts)
    
    entropias_1_barabasi.append(entropia_1_barabasi)
    entropias_1_watts.append(entropia_1_watts)
    entropias_2_barabasi.append(entropia_2_barabasi)
    entropias_2_watts.append(entropia_2_watts)
    GPs_barabasi.append(GP_barabasi)
    GPs_watts.append(GP_watts)
    CCs_barabasi.append(CC_barabasi)
    CCs_watts.append(CC_watts)
    
fin = time.time()

print(f"El tiempo de ejecucion para obtener las medidas es: {fin - inicio}")

```

```python
fig,(ax1,ax2) = plt.subplots(1,2,figsize=(14,6))
nodos_atacados = [i/N for i in range(N+1)]

ax1.plot(nodos_atacados,entropias_1_barabasi,label='Entropia normalizada en el modelo de BA',color='red',markersize=1)
ax1.plot(nodos_atacados,entropias_1_watts,label='Entropia normalizada en el modelo de WS',color='blue',markersize=1)
ax1.plot(nodos_atacados,GPs_barabasi,label='Grado promedio en el modelo de BA',color='yellow',markersize=1)
ax1.plot(nodos_atacados,GPs_watts,label='Grado promedio en el modelo de BA',color='green',markersize=1)
ax1.set_title('Entropia vs Grado Promedio cuando las redes tienen fallos')
ax1.set_xlabel('% de nodos atacados')
ax1.set_ylabel('medidas de robustez')
ax1.grid(True)
ax1.legend(title='Medidas y Modelo de Redes')

ax2.plot(nodos_atacados,entropias_1_barabasi,label='Entropia normalizada en el modelo de BA',color='red',markersize=1)
ax2.plot(nodos_atacados,entropias_1_watts,label='Entropia normalizada en el modelo de WS',color='blue',markersize=1)
ax2.plot(nodos_atacados,CCs_barabasi,label='Componentes conexas en el modelo de BA',color='yellow',markersize=1)
ax2.plot(nodos_atacados,CCs_watts,label='Componentes conexas en el modelo de WS',color='green',markersize=1)
ax2.set_title('Entropia vs Componentes Conexas cuando las redes tienen fallos')
ax2.set_xlabel('% de nodos atacados')
ax2.set_ylabel('medidas de robustez')
ax2.grid(True)
ax2.legend(title='Medidas y Modelo de Redes')
```
