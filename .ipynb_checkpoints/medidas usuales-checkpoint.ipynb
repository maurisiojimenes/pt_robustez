```python
import graficas as g
import matplotlib.pyplot as plt
import time
import networkx as nx
from PIL import Image
import os
import random
```

```python
print(os.getcwd())
```

```python
G_BA = g.crear_grafica('barabasi',500,10,0)
G_WS = g.crear_grafica('watts',500,10,0.5)

N = len(G_BA)

AD_BA = [g.grado_promedio(G_BA)]
AD_WS = [g.grado_promedio(G_WS)]

inicio = time.time()

for i in range(N):
    #nx.draw(G_BA)
    #plt.show()
    G_BA = g.ataques(G_BA,'degree')
    G_WS = g.ataques(G_WS,'degree')
    
    AD_BA.append(g.grado_promedio(G_BA))
    AD_WS.append(g.grado_promedio(G_WS))

fin = time.time()

print(f"El tiempo para obtener los AD de ambas redes fue: {fin - inicio}")
```

```python
plt.figure(figsize=(10,8))
nodos_atacados = [i/N for i in range(N+1)]

plt.plot(nodos_atacados,AD_BA,label='Grado Promedio en el modelo de BA',marker='o',markersize=1,color='red')
plt.plot(nodos_atacados,AD_WS,label='Grado Promedio en el modelo de WS', marker='x',markersize=1,color='blue')
plt.title("Gráfica Del Grado Promedio Cuando La Red Tiene Fallos")
plt.xlabel('Porcentaje De Nodos Atacados')
#plt.xscale('log')
#plt.yscale('log')
plt.ylabel('Grado Promedio')
plt.grid(True)
plt.legend(title='Modelos De Redes',loc='best')
```

```python
G = g.crear_grafica('watts',20,3,1)
nodo_mayor_grado = max(G.degree, key=lambda x: x[1])[0]

colores = ["red" if nodo == nodo_mayor_grado else "lightblue" for nodo in G.nodes()]

plt.figure(figsize=(8, 6))
nx.draw(G, with_labels=True, node_color=colores, node_size=500, font_weight="bold")
plt.title(f"Nodo de mayor grado: {nodo_mayor_grado}")
plt.show()
```

```python
G = g.crear_grafica('watts',35,5,1)
nodo_elegido = random.choice(list(G.nodes()))
nodo_grado_mayor = max(G.degree(), key=lambda x:x[1])[0]

colores = ['red' if nodo == nodo_elegido else 'lightyellow' if nodo == nodo_grado_mayor 
           else 'lightblue' for nodo in list(G.nodes())]

plt.figure(figsize=(10,6))
nx.draw(G,with_labels=True,node_color=colores,node_size=500,font_weight='bold')
plt.title(f"El nodo elegido a fallar es {nodo_elegido} y el nodo de mayor grado es {nodo_grado_mayor} con grado {G.degree(nodo_grado_mayor)}")
plt.show()
```
