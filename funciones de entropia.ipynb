```python
G_barabasi = g.crear_grafica('barabasi',1000,15,0)
G_watts = g.crear_grafica('watts',1000,15,0.5)

entropia_barabasi = en.entropia_normalizada(G_barabasi)
entropia_watts = en.entropia_normalizada(G_watts)

N = len(G_barabasi)

entropias_barabasi = [entropia_barabasi]
entropias_watts = [entropia_watts]


inicio = time.time()

for i in range(1,N,1):
    
    G_barabasi = g.ataques(G_barabasi,'random')
    G_watts = g.ataques(G_watts,'random')
    
    entropia_barabasi = en.entropia_normalizada(G_barabasi)
    entropia_watts = en.entropia_normalizada(G_watts)

    
    entropias_barabasi.append(entropia_barabasi)
    entropias_watts.append(entropia_watts)
    
    
fin = time.time()
print(f"el tiempo de ejecucion para obtener las entropias de la red fue: {fin - inicio}")
```
