```python
G_BA = g.crear_grafica('barabasi',500,10,0)
G_WS = g.crear_grafica('watts',500,10,0.5)

N = len(G_BA)

AD_BA = [g.grado_promedio(G_BA)]
AD_WS = [g.grado_promedio(G_WS)]

inicio = time.time()

for i in range(N):
    G_BA = g.ataques(G_BA,'degree')
    G_WS = g.ataques(G_WS,'degree')
    
    AD_BA.append(g.grado_promedio(G_BA))
    AD_WS.append(g.grado_promedio(G_WS))

fin = time.time()

print(f"El tiempo para obtener los AD de ambas redes fue: {fin - inicio}")
```

```python
plt.figure(figsize=(10,8))
nodos_atacados = [i/N for i in range(N+1)]

plt.plot(nodos_atacados,AD_BA,label='Grado Promedio en el modelo de BA',marker='o',markersize=1,color='red')
plt.plot(nodos_atacados,AD_WS,label='Grado Promedio en el modelo de WS', marker='x',markersize=1,color='blue')
plt.title("Gráfica Del Grado Promedio Cuando La Red Tiene Fallos En Escala Logarítmica")
plt.xlabel('Porcentaje De Nodos Atacados')
plt.xscale('log')
plt.yscale('log')
plt.ylabel('Grado Promedio')
plt.grid(True)
plt.legend(title='Modelos De Redes',loc='best')
```

```python
G_BA = g.crear_grafica('barabasi',500,15,0)
G_WS = g.crear_grafica('watts',500,15,0.5)
```

```python
indices_robustez_BA_1 = []
indices_robustez_WS_1 = []
indices_robustez_BA_2 = []
indices_robustez_WS_2 = []

inicio = time.time()
for i in range(1,len(G_BA)):
    print(f"Iteración: {i}")
    G_1 = G_BA.copy()
    G_2 = G_WS.copy()
    G_3 = G_BA.copy()
    G_4 = G_WS.copy()
    
    indice_robustez_BA_1 = g.indice_robustez(G_1,g.LCC,1,'random',i)
    indice_robustez_WS_1 = g.indice_robustez(G_2,g.LCC,1,'random',i)
    indice_robustez_BA_2 = g.indice_robustez(G_3,g.LCC,2,'random',i)
    indice_robustez_WS_2 = g.indice_robustez(G_4,g.LCC,2,'random',i)
    
    indices_robustez_BA_1.append(indice_robustez_BA_1)
    indices_robustez_WS_1.append(indice_robustez_WS_1)
    indices_robustez_BA_2.append(indice_robustez_BA_2)
    indices_robustez_WS_2.append(indice_robustez_WS_2)
    
final = time.time()
print(f"El tiempo de ejecución es: {final - inicio}")
```

```python
k = [i for i in range(1,len(G_BA))]

plt.figure(figsize=(10,6))
plt.plot(k,indices_robustez_BA_1, label = 'Indice de robustez de red de BA con w = 1/N',color='blue',markersize=1)
plt.plot(k,indices_robustez_WS_1, label = 'Indice de robustez de red de WS con w = 1/N',color='red',markersize=1)
plt.plot(k,indices_robustez_BA_2, label = 'Indice de robustez de red de BA con w = 1/LCC',color='pink',markersize=1)
plt.plot(k,indices_robustez_WS_2, label = 'Indice de robustez de red de WS con w = 1/LCC',color='orange',markersize=1)

plt.title('Gráfica de los índices de robustez usando el LCC cuándo la red tiene un escenario de fallos')

#plt.yscale('log')
#plt.xscale('log')

plt.xlabel('k')
plt.ylabel('Índices de robustez')
plt.legend(title='Modelos de redes y peso usado para los índices')
plt.grid(True)
plt.show()
```
