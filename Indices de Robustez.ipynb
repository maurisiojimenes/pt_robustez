```python
G_BA = g.crear_grafica('barabasi',100,5,0)
G_WS = g.crear_grafica('watts',100,10,0.5)
```

```python
indices_robustez_BA = []
indices_robustez_WS = []

N = len(G_BA)

for i in range(1,N,1):
    print(f"iteración: {i}")
    G_1 = G_BA.copy()
    G_2 = G_WS.copy()
    indice_1= g.indice_robustez(G_1,en.entropia,2,'degree',i)
    indice_2= g.indice_robustez(G_2,en.entropia,2,'degree',i)
    indices_robustez_BA.append(indice_1)
    indices_robustez_WS.append(indice_2)
    
```

```python
fig,axs = plt.subplots(1,1,figsize=(12,6))

axs.plot(indices_robustez_BA,label='modelo - BA',color='skyblue')

axs.plot(indices_robustez_WS,label='modelo - WS',color='darkred')
axs.set_title('Indies de robustez de entropia de Shanon en escenario de fallos')
axs.grid(True)
axs.legend(loc='best')
```
