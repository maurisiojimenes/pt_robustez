```python
G_original = g.crear_grafica('barabasi',500,10,0)

N = len(G_original)
plt.figure(figsize=(10, 6))

indices_robustez = []

inicio = time.time()

for i in range(N):
    G = G_original.copy()
    
    R = g.indice_robustez(G,en.entropia_distribucion_1,'constante','random',i)
        
    indices_robustez.append(R)

fin = time.time()

print(f"El tiempo que tomo calcular los indices de robustez fue: {fin - inicio}")

k = [i for i in range(0,500,1)]

plt.plot(k,indices_robustez,label='Indices de robustez en una red de BA',color='blue',markersize=1)
plt.title('Gr√°fica de los indices de robustez atacando k nodos')
plt.xlabel('k')
plt.ylabel('indices de robustez')
plt.grid(True)
plt.show()
```
