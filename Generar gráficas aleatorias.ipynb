```python
import matplotlib.pyplot as plt
import graficas as g
```

```python
for i in range(1,11,1):
    p = 0.1*i
    print(p)
    G_1 = g.crear_grafica('watts',50,2,p)
    G_2 = g.crear_grafica('watts',250,15,p)
    G_3 = g.crear_grafica('watts',500,10,p)
    G_4 = g.crear_grafica('watts',1000,15,p)


    distribucion_grados_1 = g.obtener_distribucion_grados(G_1)
    grados_1 = [grado for grado,_ in distribucion_grados_1]
    frecuencias_1 = [frecuencia*len(G_1) for _,frecuencia in distribucion_grados_1]

    distribucion_grados_2 = g.obtener_distribucion_grados(G_2)
    grados_2 = [grado for grado,_ in distribucion_grados_2]
    frecuencias_2 = [frecuencia*len(G_2) for _,frecuencia in distribucion_grados_2]

    distribucion_grados_3 = g.obtener_distribucion_grados(G_3)
    grados_3 = [grado for grado,_ in distribucion_grados_3]
    frecuencias_3 = [frecuencia*len(G_3) for _,frecuencia in distribucion_grados_3]

    distribucion_grados_4 = g.obtener_distribucion_grados(G_4)
    grados_4 = [grado for grado,_ in distribucion_grados_4]
    frecuencias_4 = [frecuencia*len(G_4) for _,frecuencia in distribucion_grados_4]

    # Crear figura y ejes (2x2)
    fig, axs = plt.subplots(2, 2, figsize=(12, 12))

    # Primer subplot
    axs[0, 0].bar(grados_1, frecuencias_1, color='skyblue')
    axs[0, 0].set_title(f'Distribución de grados (n=50, m=2, p={p})')
    axs[0, 0].set_xlabel('Grado')
    axs[0, 0].set_ylabel('Frecuencia')

    # Segundo subplot
    axs[0, 1].bar(grados_2, frecuencias_2, color='salmon')
    axs[0, 1].set_title(f'Distribución de grados (n=250, m=5, p={p})')
    axs[0, 1].set_xlabel('Grado')
    axs[0, 1].set_ylabel('Frecuencia')

    # Tercer subplot
    axs[1, 0].bar(grados_3, frecuencias_3, color='mediumseagreen')
    axs[1, 0].set_title(f'Distribución de grados (n=500, m=10,p={p})')
    axs[1, 0].set_xlabel('Grado')
    axs[1, 0].set_ylabel('Frecuencia')

    # Cuarto subplot
    axs[1, 1].bar(grados_4, frecuencias_4, color='orchid')
    axs[1, 1].set_title(f'Distribución de grados (n=1000, m=15,p={p})')
    axs[1, 1].set_xlabel('Grado')
    axs[1, 1].set_ylabel('Frecuencia')

    # Ajustar el espaciado
    plt.tight_layout()
    plt.show()
```
