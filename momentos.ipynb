```python
G_barabasi = g.crear_grafica('barabasi',1000,100,0)
G_watts = g.crear_grafica('watts',1000,100,0.5)
N = len(G_barabasi)
plt.figure(figsize=(10, 6))
momentos_barabasi = [g.momentos(G_barabasi,1)]
momentos_watts = [g.momentos(G_watts,1)]
for _ in range(len(G_barabasi)):
    G_barabasi = g.ataques(G_barabasi,'random')
    G_watts = g.ataques(G_watts,'random')
    momentos_barabasi.append(g.momentos(G_barabasi,1))
    momentos_watts.append(g.momentos(G_watts,1))
nodos_atacados = [i/N for i in range(N)]
nodos_atacados.insert(0,0)

# graficamos los resultados:
plt.plot(nodos_atacados, momentos_barabasi, label="Barabási", color='blue', marker='o', linestyle='-', markersize=1)
plt.plot(nodos_atacados, momentos_watts, label="W-S", color='red', marker='x', linestyle='-', markersize=1)

plt.title("Gráfica del primer momento atacado por criterio aleatorio")
plt.xlabel("% de nodos atacados")
plt.ylabel("<k>")

plt.legend(title="Modelos de Redes", loc='best')
plt.grid(True)
plt.show()
```
